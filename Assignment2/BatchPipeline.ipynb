```python
from pyspark.sql import SparkSession
from pyspark import SparkConf

sparkConf = SparkConf()
sparkConf.setMaster("spark://spark-master:7077")
sparkConf.setAppName("GCSExample")
sparkConf.set("spark.driver.memory", "2g")
sparkConf.set("spark.executor.cores", "1")
sparkConf.set("spark.driver.cores", "1")
# create the spark session, which is the entry point to Spark SQL engine.
spark = SparkSession.builder.config(conf=sparkConf).getOrCreate()

# Setup hadoop fs configuration for schema gs://
conf = spark.sparkContext._jsc.hadoopConfiguration()
conf.set("fs.gs.impl", "com.google.cloud.hadoop.fs.gcs.GoogleHadoopFileSystem")
conf.set("fs.AbstractFileSystem.gs.impl", "com.google.cloud.hadoop.fs.gcs.GoogleHadoopFS")

#  Google Storage File Path
gsc_file_path = 'gs://dejads_input_assignment2/' 
# Create data frame
athletes_df = spark.read.format("csv").option("header", "true") \
       .load(gsc_file_path+'Athletes.csv')
athletes_df.printSchema()
athletes_df.show()

teams_df = spark.read.format("csv").option("header", "true") \
       .load(gsc_file_path+'Teams.csv')
teams_df.printSchema()
teams_df.show()

medals_df = spark.read.format("csv").option("header", "true") \
       .load(gsc_file_path+'Medals.csv')
medals_df.printSchema()
medals_df.show()
```

```python
# Stop the spark context
spark.stop()
```
